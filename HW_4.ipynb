4\. **Trigonometric functions**

Use `np.linspace` to create an array of 100 numbers between $0$ and $2\pi$ (inclusive).

  * Extract every 10th element using the slice notation
  * Reverse the array using the slice notation
  * Extract elements where the absolute difference between the `sin` and `cos` functions evaluated for that element is $< 0.1$
  * **Optional**: make a plot showing the sin and cos functions and indicate where they are close

````python
import cmath
import matplotlib.pyplot as plt
trig_array=np.linspace(0,2*cmath.pi,100)#creating the array
print("Array is:\n",trig_array)
print("every 10th element is:\n",trig_array[::10])#extract every 10th element
print("reverse of the array:\n",trig_array[::-1])#reverse the array
#Extract elements where the absolute difference between the sin and cos functions evaluated for that element is  <0.1
mask_trig=(abs(np.sin(trig_array) - np.cos(trig_array)) < 0.1)
extracted_matrix= trig_array[mask_trig]
print("abs diff matrix:\n",extracted_matrix)
#plot showing the sin and cos functions and indicate where they are close
x = trig_array
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,x,z)
plt.title('Plot')
plt.legend(['sin(x)', 'cos(x)'])
plt.show()
````

6\. **Broadcasting**

Use broadcasting to create a grid of distances.

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles.

The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Build a 2D grid of distances among each city along Route 66
  * Convert the distances in km

````python
import pandas as pd
position=np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])
distance_grid=np.abs(position - position[:,np.newaxis]) #np.newaxis to add multiple dimensions
# it will keep shifting dimensions and computing
df=pd.DataFrame(distance_grid) # 2D grid
cities=["Chicago", "Springfield", "Saint-Louis", "Tulsa", "Oklahoma City", "Amarillo", "Santa Fe",
        "Albuquerque", "Flagstaff", "Los Angeles"]
df.columns = cities
df.index = cities
df
conversion_factor = 0.62137119 # miles to km conversion
miles_to_km = np.array(np.divide(position,conversion_factor).round(decimals=1))
print (miles_to_km,"\n")
distance_grid=np.abs(miles_to_km - miles_to_km[:,np.newaxis])
df=pd.DataFrame(distance_grid) # 2D grid
cities=["Chicago", "Springfield", "Saint-Louis", "Tulsa", "Oklahoma City", "Amarillo", "Santa Fe",
        "Albuquerque", "Flagstaff", "Los Angeles"]
df.columns = cities
df.index = cities
print("2D grid of distances among each city along Route66 in Km")
df
````

7\. **Prime numbers sieve**

Compute the prime numbers in the 0-N (start with N=99) range with a sieve (mask).

  * Constract a shape (N,) boolean array, which is the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (with `timeit`); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
import math
print ("Enter a number:\n")
num = int(input())
prime_list=[]
for i in range(2,num+1):
    prime_list.append(i)#we append all the numbers first and then we seed the others out
        
i=2
while(i<=math.sqrt(num)):# to seed out different multiples of numbers until sqrt(number)
    if i in prime_list: # if i is in the list(it will be)we have to delete its multiples
        for j in range(i*2,num+1,i): # finding the multiples
            if j in prime_list:
                prime_list.remove(j)#removes the multiples
    
    i=i+1
        
print(prime_list)
````
